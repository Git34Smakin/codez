```python
# --- Data Cleaning and Preprocessing ---

import numpy as np
import pandas as pd # Import pandas

# Load data into a DataFrame
# Assuming your data is in a CSV file named 'data.csv'
try:
    df = pd.read_csv('mydataset.csv')
except FileNotFoundError:
    print("Error: 'data.csv' not found. Please make sure the data file is in the correct location.")
    # You might want to add an exit or raise an exception here if the file is essential
    exit()


# 1. Replace "?" with NaN
df.replace("?", np.nan, inplace=True)

# 2. Drop duplicates
df.drop_duplicates(inplace=True)

# 3. Convert numerical columns from object to float
numeric_cols = ["normalized-losses", "bore", "stroke",
                "horsepower", "peak-rpm", "car-price"]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# 4. Handle missing values
# Drop rows where car-price is missing (target variable)
df.dropna(subset=["car-price"], inplace=True)

# Impute numerical columns with mean
for col in ["normalized-losses", "bore", "stroke", "horsepower", "peak-rpm"]:
    df[col].fillna(df[col].astype(float).mean(), inplace=True)

# Impute categorical columns with mode
for col in ["num-of-doors"]:
    df[col].fillna(df[col].mode()[0], inplace=True)

# 5. Check cleaned dataset info
print("After preprocessing:")
print(df.info())
print("Number of missing values per column:")
print(df.isnull().sum())
```
