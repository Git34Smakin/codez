```python
import pandas as pd

# Define the path to your dataset
file_path = "/content/drive/MyDrive/dataset/Car_Sales.csv"

# Load the dataset
car_data = pd.read_csv(file_path)

# Display first 5 rows
car_data.head()
```

```python
# Shape of dataset (rows, columns)
print("Dataset Shape:", car_data.shape)

# Column names
print("\nColumns in dataset:")
print(car_data.columns)

# General info (data types, non-null values)
print("\nDataset Info:")
print(car_data.info())

# Summary statistics (numerical features)
print("\nSummary Statistics:")
print(car_data.describe())

# Check missing values
print("\nMissing Values:")
print(car_data.isnull().sum())

# Check first few rows
print("\nFirst 5 rows:")
print(car_data.head())
```

```python
# If there's a target column like 'Sales_Category', check class distribution
if 'Sales_Category' in car_data.columns:
    print("\nClass distribution:")
    print(car_data['Sales_Category'].value_counts())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numerical columns for correlation heatmap
numerical_car_data = car_data.select_dtypes(include=['float64', 'int64'])

# Correlation heatmap
plt.figure(figsize=(10,6))
sns.heatmap(numerical_car_data.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
